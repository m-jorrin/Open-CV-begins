### Draw a line

To draw a line it is possible to use `cv2`'s method `line`. 

`cv2.line(image, (start coordinate, finish coordinate), color, width);`

<h3>Calcula:</h3>
<p>Dibuja una linea en la foto y representa la imagen original y la que has pintado al lado</p>

```python
image = cv2.imread('../imgs/windows.png')
print(image.shape)
image_new = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_new = cv2.resize(image_new, (250, 250))
l=[]
# Generate multiple image clones of the image in a list

for i in range(3):
    l.append(image_new.copy())
    width, height, dimension = 250,250,3
    line_width = 10

# check line_width
# print("line width:",line_width)

image_d = cv2.resize(image_new, (240, 250))
cv2.line(image_d, (0,0),(250,250),(250,250,250),line_width)

cv2.line(image_d.copy(), (0,250),(250,0),(0,250,250),line_width)

# If you want to see output and check that there are 250 250 250 value pixels, delete ;

fig, axs = plt.subplots(1, 2, figsize=(15, 15))
axs[0].imshow(image_new)
axs[0].set_title("Original img")
axs[1].imshow(image_d, cmap='gray')
axs[1].set_title("Drawn img")

plt.show()
```

### Draw a circle

To draw a circle it is possible to use `cv2`'s method `circle`. 

`cv2.circle(image, centre_coordinate, radius, color, width)`
**_note_**: If `width` value is negative, then the circle if filled: (*negative == fill*)

<h3>Calcula:</h3>
<p>Dibuja un circulo en la imagen y representa todas, la original, una con el circulo relleno y otra sin rellenar.</p>

```python
image = cv2.imread('../imgs/windows.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


# Generate multiple image clones of the image in a list
photos = [image.copy() for i in range(3)]
width, height, dimension = 250, 250, 3
line_width = int((width*height)/10000)

# check line_width
print("line width:",line_width)

 # If you want to see output and check that there are 250 250 250 value pixels, delete ;
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
photos[0]=cv2.resize(photos[0], (250,250))
axs[0].imshow(photos[0])
axs[0].set_title("Original img")

photos[1]=cv2.resize(photos[1], (250,250))
photos[1]=cv2.circle(photos[1], (125,125), 100, (250,250,250), line_width)

axs[1].imshow(photos[1])
axs[1].set_title("Not filled circle")

photos[2]=cv2.resize(photos[2], (250,250))
photos[2]=cv2.circle(photos[2], (125,125), 50, (250,250,250), line_width*-1)


axs[2].imshow(photos[2])
axs[2].set_title("FIlled circle")
plt.show()
```

### Draw a rectangle

To draw a rectangle it is possible to use `cv2`'s method `rectangle`.

`cv2.rectangle(image, border_coor_1, border_coor_2, color, width)`

<h3>Calcula:</h3>
<p>Dibuja un rectangulo en la imagen y representa todas, la original, una con el rectangulo relleno y otra sin rellenar.</p>

```python
image = cv2.imread('../imgs/windows.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Generate multiple image clones of the image in a list
photos = [image.copy() for i in range(3)]
width, height, dimension = 250, 250, 3
line_width = int((width*height)/10000)
for i in range(len(photos)):
    photos[i] =cv2.resize(photos[i], (250,250))

# check line_width
print("line width:",line_width)

 # If you want to see output and check that there are 250 250 250 value pixels, delete ;
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
axs[0].imshow(photos[0])
axs[0].set_title("Original img")

photos[1]=cv2.rectangle(photos[1], (0,100), (300,120), (250,250,250), line_width)
axs[1].imshow(photos[1])
axs[1].set_title("Not filled rectangle")

photos[2]=cv2.rectangle(photos[2], (50,50), (200,200), (250,250,250), line_width*-1)
axs[2].imshow(photos[2])
axs[2].set_title("Filled rectangle")
plt.show()
```

## Wrap up

Following points need to be clear in your mind to be able to advance in `cv2`:

* Create a copy of the image in order to not rewrite the original:
    * *idea:* list of images.
    
* Reload images: 
    *  Reset list of images.
   
* Draw on images:
    *  Use relative drawing so it is simple to refactor when you change images.
    *  OpenCV has different shapes to draw including text.
