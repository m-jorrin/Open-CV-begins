```python
#Get the img matrix
image = cv2.imread('../imgs/perro.jpeg')
print(image.shape)
#Convert the image to gray
image_gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# plt.imshow(image_gray, cmap='Dark2') Dark2 psicodelia pura
# plt.imshow(image_gray, cmap='BuPu_r') # el _r es para no invertir colores
# plt.imshow(image_gray, cmap='Greys_r')
# plt.show()
plt.imshow(image_gray, cmap='gray')
plt.show()

#Import directly grayscale image
direct_image_grey=cv2.imread('../imgs/perro.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(direct_image_grey, cmap='gray')
plt.show()
```

La imagen directamente convertida devuelve unos valores m√°s contrastados de blanco respecto al gris en esta imagen. 

La diferencia entre ambas es que los valores que convierte no corresponden a los reales, la funcion cv2.COLOR_BGR2GRAY interpreta los valores de la imagen en formato BGR y no en el original real que es RGB. Como s√≠ ocurre al importarse directamente al introducir en el m√©todo .imread() el par√°metro 'cv2.IMREAD_GRAYSCALE'.

```python
#Change BGR to RGB 

#Set subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
#Original image on RGB
image_RGB = cv2.imread('../imgs/perro.jpeg')
image_RGB = cv2.cvtColor(image_RGB, cv2.COLOR_BGR2RGB)
image_directly = cv2.imread('../imgs/perro.jpeg', cv2.IMREAD_GRAYSCALE)
axs[0].imshow(image_RGB)
axs[0].set_title("Original img")
#Bad image on BGR
bad_image=cv2.imread('../imgs/perro.jpeg')
gray_image = cv2.cvtColor(bad_image, cv2.COLOR_BGR2GRAY)
axs[1].imshow(gray_image, cmap="gray")
axs[1].set_title("Converted to gray")
#Good image converted to gray from RGB
axs[2].imshow(image_directly, cmap="gray")
axs[2].set_title("Directly converted to gray")
plt.show()
```

## Write the images

In order to write the images is important to note that `cv2` method `imwrite` changes the input image to RGB as `cv2` thinks most of time it is better to write this way.

So in order to write the images correctly It is necessary to have the input image as BGR.

<h3>Calcula:</h3>
<p>Guarda las 3 imagenes utilizando imwrite.</p>

```python

# Convert to BGR
image_RGB_BGR = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2BGR)
gray_image__from_grays_to_BGR =cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)
goodgray_image__from_gray_to_BGR=cv2.cvtColor(image_directly, cv2.COLOR_GRAY2BGR)

carpeta_edits = "/home/mjorrinb/Documentos/SIALI/github/Open-CV-begins/imgs_mods"
nombre_archivo = "original_RGB.jpg"
#
#       A la hora de guardar las imagenes se ha de convertir primero a formato RGB que es el que interpreta openCV 
#       aunque posteriormente lo transforme a formato RGB.
#
# Save images
cv2.imwrite(os.path.join(carpeta_edits, 'original_RGB.jpg'), image_RGB_BGR)
cv2.imwrite(os.path.join(carpeta_edits, 'mala_gris.jpg'), gray_image__from_grays_to_BGR)
cv2.imwrite(os.path.join(carpeta_edits, 'buena_gris.jpg'), goodgray_image__from_gray_to_BGR)

# print images
fig, axs = plt.subplots(1, 3, figsize=(15, 15))
axs[0].imshow(image_RGB_BGR)
axs[0].set_title("Original img de RGB a BGR")
axs[1].imshow(gray_image__from_grays_to_BGR, cmap="gray")
axs[1].set_title("Convertet to gray")
axs[2].imshow(goodgray_image__from_gray_to_BGR, cmap="gray")
axs[2].set_title("Directly convertet to gray")
plt.show()
```

## Wrap up

Following points need to be clear in your mind to be able to advance in `cv2`:

* Importing images:
    *  Convert from BGR to RGB.
    
* Process images: 
    *  Do it with RGB image.
   
* Write images:
    *  Convert RGB to BGR in order to write it (`cv2` automatically converts BGR to RGB).
